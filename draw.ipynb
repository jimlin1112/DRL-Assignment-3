```python
import gymnasium as gym
import os
import matplotlib.pyplot as plt
import imageio
import numpy as np
from IPython.display import Image
import torch
import random
from collections import deque

import torch.nn.functional as F
import torch.nn as nn
from typing import Dict, List, Tuple

import re
import matplotlib.pyplot as plt

# 用於存放解析後的 steps 和 rewards
a1 = []
reward = []

# 正則表達式：匹配 Reward 和 Step
pattern = re.compile(r'Step\s*(\d+).*?Reward\s*([\d.]+)')

# 打開並逐行讀取日誌檔
with open('output.txt', 'r') as f:
    for line in f:
        match = pattern.search(line)
        if match:
            tmp1 = float(match.group(1))
            tmp2 = float(match.group(2))
            a1.append(tmp1)
            reward.append(tmp2)

reward_avg = []
avg_num = 1000
for i in range(avg_num, len(reward)):
    reward_avg.append(np.mean(reward[i-avg_num:i]))

# 繪圖
plt.figure(figsize=(20, 10))
plt.plot(a1[avg_num:], reward_avg, marker='o', linestyle='-')
plt.grid(True)
plt.tight_layout()
plt.show()
```
